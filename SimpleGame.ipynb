The Hunger Games APA citation: Collins, S. (2008). The Hunger Games. First edition. New York, Scholastic Press.

```python
#!/usr/bin/python
import random

class Player:
    """Class to represent the player, with attributes that affect gameplay."""
    def __init__(self):
        # Initialize player with health and random attributes for strength, agility, and intelligence.
        self.health = 100
        self.strength = random.randint(5, 10)
        self.agility = random.randint(5, 10)
        self.intelligence = random.randint(5, 10)
        self.supplies = []

    def display_stats(self):
        """Display player's current statistics."""
        print(f"Health: {self.health}, Strength: {self.strength}, Agility: {self.agility}, Intelligence: {self.intelligence}")

player = Player()

def win():
    """Function to call when the player wins the game."""
    print("You survived... at least, the first day of the Hunger Games")

def fail():
    """Function to call when the player fails the game."""
    print("Your vision starts to fade, as you begin to drift to sleep. You hear a muffled cannon sound in the distance.")

def get_supplies(count):
    """Allows the player to choose supplies up to a specified count."""
    available_supplies = ["SWORD", "BOW", "ARROWS", "CLUB", "KNIFE", "BACKPACK", "WATER"]
    random.shuffle(available_supplies)  # Shuffle supplies to randomize the availability order.
    print("Available supplies:", available_supplies)
    for i in range(count):
        if not available_supplies:
            break
        supply = input("Type the supply you'd like to get or STOP when you are done grabbing supplies\n >").upper()
        if supply == "STOP":
            break
        elif supply in available_supplies:
            player.supplies.append(supply)
            available_supplies.remove(supply)
            print(f"You have obtained a {supply}.")
        else:
            print(f"You couldn't find {supply} and wasted time looking for it.")

    if len(player.supplies) > count:
        print("You were too fixated on grabbing supplies to notice an enemy sneak up behind you and attack!")
        fail()
        return False
    return True

def combat():
    """Simulates combat scenario based on player's supplies and attributes."""
    # Enhance strength if player has combat-effective supplies.
    combat_strength = player.strength + 2 if "SWORD" in player.supplies or "KNIFE" in player.supplies else player.strength

    enemy_strength = random.randint(3, 10)
    print("An enemy is approaching!")
    choice = input("Do you FIGHT or RUN? ").upper()
    if choice == "FIGHT":
        if combat_strength > enemy_strength:
            print("You successfully defeated the enemy.")
            win()
        else:
            print("The enemy was too strong. You were defeated.")
            fail()
    elif choice == "RUN" and player.agility > 5:
        print("You managed to escape.")
        win()
    else:
        print("You couldn't escape.")
        fail()

def game():
    """Main function to run the game, presenting choices and initiating game phases."""
    player.display_stats()
    opening = ("You have been selected to participate in the Hunger Games. Many believe you will not survive the first day. Can you survive? "
               "To play, type in the FULLY CAPITALIZED words as your options for actions. \nYou find yourself on a platform rising up from the ground, "
               "and as the first ray of sunlight reaches you, you hear a countdown. \n 10, 9, 8... \nYou are in the Hunger Games, where you must fight for your survival \n 7, 6, 5... \nYou know you will only have a couple of seconds to decide 3, 2, 1...")
    print(opening)
    choices = {"FIGHT": "You rush to the center arena where various supplies are scattered about.",
               "HIDE": "You hide in the outskirts, watching the chaos from a distance.",
               "SURRENDER": "You step off the platform prematurely, triggering an explosion."}

    choice = input("Choose FIGHT, HIDE, or SURRENDER: ").upper()
    if choice in choices:
        print(choices[choice])
        if choice == "SURRENDER":
            fail()
        else:
            if get_supplies(3 if choice == "FIGHT" else 5):
                combat()
    else:
        print("Invalid choice. Please restart the game and select a valid option.")

if __name__ == "__main__":
    game()
```
