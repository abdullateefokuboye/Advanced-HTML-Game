```python
#!/usr/bin/python

def win():
    print("You survived... at least, the first day of the Hunger Games")

def fail():
    print("Your vision starts to fade, as you begin to drift to sleep. You hear a muffled cannon sound in the distance.")
    
def get_supplies(count):
    supplies = ["SWORD", "BOW", "ARROWS", "CLUB", "KNIFE", "BACKPACK", "WATER"]
    waitingForSupplies = True
    counter = count
    for i in range(counter):
        print(supplies)
        supply = input("Type the supply you'd like to get or STOP when you are done grabbing supplies\n >")
        supply = supply.upper()
        if supply == "STOP":
            break
        try:
            supplies.remove(supply)
        except:
            print("You couldn't find ", supply, "and wasted time looking for it")

    else:
        print("You were too fixated on grabbing supplies to notice an enemy sneak up behind you and attack!")
        fail()
        return False
    return True

def game():
    opening = "You have been selected to participate in the Hunger Games. Many believe you will not survive the first day. Can you survive? To play, type in the FULLY CAPITALIZED words as your options for actions \n You find yourself on a platform rising up from the ground, and as the first ray of sunlight reaches you, you hear a count down. \n 10, 9, 8\
\n You are in the Hunger Games, where you must fight for your survival \n 7, 6, 5 \n You know you will only have a couple of seconds to decide 3,2,..."
    print(opening)
    fight = "You run to the center arena where various supplies are scattered about, you see:"    
    hide = "After hiding in the outskirts of the clearing for quite sometime, the action has calmed down. You see some bodies and scattered supplies. Amongst them you see:"
    surrender = "You step off the platform before 1, and the earth beneath you errupts in an explosion"
    befriend = "The former enemy is grateful for your decision and you team up to survive the night."
    finish = "You finish off the enemy. There are no threats around as the sun sets."
    waitingForChoice1 = True
    while(waitingForChoice1 is True):
        startChoice = input("Will you rush into the center of the arena and FIGHT, will you HIDE for your safety, or will you SURRENDER because things are too hopeless \n >")
        startChoice = startChoice.upper()
        if startChoice == "FIGHT":
            waitingForChoice1 = False
            print(fight)
            if get_supplies(3):
                choice3 = ""
                while (choice3 != "BEFRIEND" and choice3 != "FINISH"):
                    choice3 = input("You see an enemy approaching but are able to fend them off. Do you FINISH them off or try to BEFRIEND them")
                    choice3 = choice3.upper()
                    if choice3 == "BEFRIEND":
                        print(befriend)
                        win()
                    elif choice3 == "FINISH":
                        print(finish)
                        win()
                    else:
                        print("Couldn't recognize the choice. Please choose BEFRIEND or FINISH")
                
        elif startChoice == "HIDE":
            waitingForChoice1 = False
            print(hide)
            #More time since the action has died down
            if get_supplies(5):
                choice3 = ""
                while (choice3 != "BEFRIEND" and choice3 != "FINISH"):
                    choice3 = input("You see an enemy approaching but are able to fend them off. Do you FINISH them off or try to BEFRIEND them")
                    choice3 = choice3.upper()
                    if choice3 == "BEFRIEND":
                        print(befriend)
                        win()
                    elif choice3 == "FINISH":
                        print(finish)
                        win()
                    else:
                        print("Couldn't recognize the choice. Please choose BEFRIEND or FINISH")
        elif startChoice == "SURRENDER":
            waitingForChoice1 = False
            print(surrender)
            fail()
        else:
            waitingForChoice1 = True
            print("Couldn't recognize the choice. Please choose FIGHT, HIDE, or SURRENDER.")
game()
```
